```python
df = pd.read_csv('/Users/cvanwieren/Library/CloudStorage/Dropbox/GitHub/Sercon/datafiles/23_09_15_CV_OFP_AF/samples/23_09_15_CV_OFP_AF.csv')
df1 = df[df['include']==True]

stds = df1[df1['is_smpl']==False].reset_index(drop=True)
fans = df1[df1['is_smpl']==True].reset_index(drop=True)
fans
```

```python
#keys = list(i.split('D_',1)[1] for i in fans['Name'])
keys = fans['Name']
vals = ['limestone clast','light matrix','light matrix','dark matrix','thin fan','laminated muds','thick fan top','thick fan middle',
       'thick fan top','thick fan bottom','laminated muds']
facies = dict(zip(keys,vals))
facies
```

```python
fans['facies'] = fans['Name'].map(facies)
```

```python
sub = fans[~fans['facies'].str.contains('matrix')]#|clast|muds')]
fig,ax = plt.subplots(1,1,figsize=(5,4.5))

sns.scatterplot(x=sub['d18o_tp'],y=sub['d13c_tp'],edgecolor='.3',hue=sub['facies'],palette='tab10',s=90)

ax.grid()
ax.set_xlabel('$\delta^{18}$O ()')
ax.set_ylabel('$\delta^{13}$C ()')

handles, labels = ax.get_legend_handles_labels()

for ha in handles: 
    ha.set_linewidth(.5)
    ha.set_edgecolor('.3')

ax.legend(loc=3,title=False,markerscale=1.5,framealpha=1)

# label
labs = list(i.split('D_',1)[1] for i in sub['Name'])
x=sub['d18o_tp']
y=sub['d13c_tp']

for i,j,l in zip(x,y,labs):
    y_off = 0.05
    x_off = 0.01

    if (j + y_off >= ax.get_ylim()[1])|(i + x_off >= ax.get_xlim()[1]):
        y_off = -y_off
        x_off = -x_off
        
    ax.text(i+x_off,j+y_off,l,color='.3')
```
