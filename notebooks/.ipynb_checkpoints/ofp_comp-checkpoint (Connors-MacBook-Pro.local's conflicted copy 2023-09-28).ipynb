```python
def draw_rect(init,scale,y_scaled,c,ax,orient='right',buf=0.1):
    # rectangles
    wi = 0.1
    width = wi/(scale/2)
    # scaled with width of rectangle propotional to 1/scale factor as in Husson et al., 2016, EPSL
    rect = Rectangle([init,np.floor(min(y_scaled))],width=width,height=max(y_scaled)-min(y_scaled),facecolor=c,edgecolor='none',alpha=1)
    ax.add_patch(rect)
    # label
    coef = -1 if orient=='right' else 1
    buf = coef*(buf/2) if scale==1 else coef*buf
    ax.text(init+buf,min(y_scaled)-10,abs(float(scale)),color=c)
    return width
```

```python
def linreg(x,y,rnd=3):
    """
    linear regression helper function updated to use the numpy.polynomial API rather than the np.polyfit and np.polyval version (outdated)
    import numpy.polynomial.polynomial as pl
    changes: cov no longer required, args are reversed so slope is p[1] and int is p[0] instead of the opposite
    args:

    returns:
        r2a: r-squared values
        ma: slope
        b: y-intercept
        eqn: equation of linear best-fit line
        line: x and y of line for plotting

    example usage:
        r2,m,b,eqn,ln = linreg(x=df['d18o'],y=df['d13c'])
        plt.plot(ln[0],ln[1],ls='--',color='.3',zorder=10)
        plt.title(eqn)
    """
    # regression on data
    p = np.polynomial.polynomial.polyfit(x,y,1)
    Xs=np.linspace(min(x),max(x))
    fit = np.polynomial.polynomial.polyval(p, Xs)
    r2 = np.corrcoef(x,y)
    m = p[1]
    inte = p[0]
    r2a = round(r2[0][1]**2,rnd)
    ma = round(m,rnd)
    b = round(inte,rnd)
    eqn = f'r$^2$ = {r2a}\n y = {ma}x + {b}'
    line = [Xs,fit]

    return r2a, ma, b, eqn, line
```

```python
fig,ax = plt.subplot_mosaic(
    """
    FABGC
    FABGC
    """,width_ratios=[0,.8,.8,.8,1],figsize=(12.5,6),sharey=False)#,layout='constrained') # first axis grid was 0.1w
plt.subplots_adjust(wspace=0)

# rects
init = 0
ys = []
for (key,val),c in zip(reversed(comp.items()),pal):
    l = key
    base = val[0]
    scale = val[1]
    sec = val[2]
    y_scaled = scale*sec['sample_height']+base
    ys.append(y_scaled.values)
    init += 0.4 
    # geochemistry
    s = 20
    # regression for carbon oxygen
    r2,_,_,_,_ = linreg(sec['d18o_carb'],sec['d13c_carb'])
    targs = [sec['d13c_carb'],sec['d18o_carb'],np.log10(sec['Sr_mmol_molCa'])]
    for a,t in zip(['A','B','G'],targs):
        sns.scatterplot(x=t,y=y_scaled,ax=ax[a],label=l+f';   $r^{{2}}$ = {r2}',legend=False,edgecolor='none',facecolor=c,marker='o',alpha=1,s=s,clip_on=False,zorder=10)

    if key=='Railway':
        width = draw_rect(0.4,scale,y_scaled,c,ax['C'])
    else:
        _ = draw_rect(init,scale,y_scaled,c,ax['C'])

# get array of all scaled heights
ys = np.concatenate(ys)

# set lims
ax['A'].set_ylim(min(ys),max(ys)) # -32,275
ax['A'].set_xlim(left=min(secs['d13c_carb'])) # -13.5,1
ax['A'].set_xticks([-12,-8,-4,0])
ax['B'].set_xticks([-20,-18,-16,-14])
ax['B'].set_xlim(left=-22)
ax['C'].set_xlim(0,np.ceil(init+width))
ax['G'].set_xlim(right=np.log10(np.nanmax(secs['Sr_mmol_molCa'])))

# sharey
ax['B'].sharey(ax['A'])
ax['C'].sharey(ax['B'])
#ax['F'].sharey(ax['A'])
ax['G'].sharey(ax['A'])

# set Sr to log scale
ticks_x = FuncFormatter(lambda x, pos: '{0:g}'.format(round(10**x,2)))
ax['G'].xaxis.set_major_formatter(ticks_x)

cols = ['tab:orange','tab:blue','tab:red']
xl = ['Sr/Ca (mmol/mol)','$\delta^{18}$O$_{carb}$ (â€°)','$\delta^{13}$C$_{carb}$ (â€°)']
axs = ['G','B','A']
ax['F'].set_ylabel('scaled stratigraphic height',fontsize=12)
ax['C'].text(0.4,0.7,'width of bar\n1/$\propto$\nscale factor',transform=ax['C'].transAxes,va='center',ha='center',fontsize=10)
rect = Rectangle((0.25, 0.6385), 0.3, 0.125, edgecolor='.3',lw=0.75,facecolor='none',transform=ax['C'].transAxes)
ax['C'].add_patch(rect)

for a,c,l in zip(axs,cols,xl):
    ax[a].set_xlabel(l,color=c)
    ax[a].xaxis.label.set_color(c)        
    ax[a].tick_params(axis='x', colors=c)
    ax[a].spines['bottom'].set_color(c)      
    ax[a].xaxis.grid(color=c,alpha=0.2)
    ax[a].tick_params(axis='y', which='both',length=0)
    ax[a].set_ylabel(None)
    ax[a].set_yticks([])
    
# legend
leg = ax['B'].legend(loc=3,bbox_to_anchor=(-0.405,0.0125),fontsize=8,framealpha=1,ncol=1)
frame = leg.get_frame()
frame.set_edgecolor('.3')
frame.set_linewidth(.75)
frame.set_boxstyle('square')

# clean up
ax['A'].spines['right'].set_visible(False)
ax['B'].spines['left'].set_visible(False)
ax['B'].spines['right'].set_visible(False)
ax['G'].spines['left'].set_visible(False)
#ax['B'].set_yticks([])
ax['C'].axis('off')
ax['F'].set_xticks([])

# unit boundaries
wmm_base = np.floor(min(comp['McKale Creek'][1]*comp['McKale Creek'][2]['sample_height'])+comp['McKale Creek'][0])
tlm_top = np.ceil(max(comp['Wolverine'][1]*comp['Wolverine'][2]['sample_height'])+comp['Wolverine'][0])
for i in ['A','B','G']:
    ax[i].axhspan(wmm_base,ax['A'].get_ylim()[1],alpha=0.075,lw=0)
    ax[i].axhspan(0,tlm_top,alpha=0.075,lw=0,color='purple')
    ax[i].axhspan(tlm_top,wmm_base,alpha=0.1,lw=0,color='grey')
    ax[i].axhspan(ax['A'].get_ylim()[0],0,alpha=0.075,lw=0,color='C1')
el = [Patch(facecolor='lightblue', edgecolor='k',label='WMM'),
      Patch(facecolor='lightgrey', edgecolor='k',label='GSM'),
      Patch(facecolor='plum', edgecolor='k',label='TLM')]
ax['C'].legend(handles=el,loc=3,bbox_to_anchor=(0.205,0.8),frameon=False,fontsize=10,handletextpad=0.4,markerfirst=True,handlelength=0.75)



# add 607.8 Ma age to side of figure
ax['G'].scatter(x=np.log10(np.nanmax(secs['Sr_mmol_molCa'])),y=(wmm_base+tlm_top)/2,marker='D',clip_on=False,zorder=10,edgecolor='k',facecolor='r',label='607.8$\pm$4.7 Ma')
el = [matplotlib.lines.Line2D([0], [0], marker='D', label='607.8 $\pm$ 4.7 Ma (Re-Os)',mfc='r',lw=0,mec='k',ms=6)]
ax['G'].legend(handles=el,loc=3,bbox_to_anchor=(1.22,0.75),frameon=False,fontsize=10,handletextpad=0.1)

fig.savefig(fig_dir+'ofp_comp.pdf',format='pdf',bbox_inches='tight',transparent=False)
```
