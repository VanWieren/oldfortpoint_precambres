```python
# canadian provinces (https://www12.statcan.gc.ca/census-recensement/alternative_alternatif.cfm?l=eng&dispext=zip&teng=lpr_000b21a_e.zip&k=%20%20%20130596&loc=//www12.statcan.gc.ca/census-recensement/2021/geo/sip-pis/boundary-limites/files-fichiers/lpr_000b21a_e.zip)
cp = gpd.read_file('/Users/cvanwieren/Library/CloudStorage/Dropbox/GitHub/OFP/OFP/Geospatial/lpr_000b21a_e/lpr_000b21a_e.shp')
cp = cp.loc[:, 'geometry']
cpa = cp.to_crs({'proj':'longlat', 'ellps':'WGS84', 'datum':'WGS84'})
```

```python
# clip vectors 
main_extent = [-120.5, -115.5, 51, 54]

polygon = box(main_extent[0], main_extent[2], main_extent[1], main_extent[3])
ab_miettemain = ab_miette.clip(polygon)
bc_miettemain = bc_miette.clip(polygon)
bc_gogmain = bc_gog.clip(polygon)
ab_gogmain = ab_gog.clip(polygon)
bc_kazamain = bc_kaza.clip(polygon)
bc_ofpmain = bc_ofp.clip(polygon)
bc_byngmain = bc_byng.clip(polygon)
bc_ismain = bc_is.clip(polygon)
ab_elsemain = ab_else.clip(polygon)
bc_elsemain = bc_else.clip(polygon)
bc_faultsmain = bc_faults.clip(polygon)
ab_faultsmain = ab_faults.clip(polygon)
```

```python
sns.set_context('notebook')

fig = plt.figure(figsize=(12, 8)) 

# build cartopy axis
#fig,ax = plt.subplots(1,4,subplot_kw={'projection': ccrs.PlateCarree()})
ax = plt.axes(projection=ccrs.PlateCarree())

def main_map(ax,extent,subplots=False):
    """
    
    """
    if subplots:
        ax.set_ylim(extent[2],extent[3])
        ax.set_xlim(extent[0],extent[1])
    else:
        # set map extent
        ax.set_extent(extent,crs=ccrs.PlateCarree())

    # add features
    ax.add_feature(prov, linestyle='--', linewidth=1, edgecolor='k', zorder=100,label='Provincial Border')
    ax.add_feature(cfeature.LAKES, linestyle='-', linewidth=0, edgecolor='k', zorder=110,label='lakes')
    ax.add_feature(cfeature.RIVERS,linestyle='-', linewidth=1, zorder=90,label='rivers')

    # add tick marks and gridlines
    #ax.set_xticks([-120,-119,-118,-117,-116], crs=ccrs.PlateCarree())
    #ax.set_yticks([51.5,52,52.5,53,53.5], crs=ccrs.PlateCarree())
    gl = ax.gridlines(draw_labels=True, lw=1.2,color='k', zorder=1000,alpha=0.3)
    gl.top_labels = False
    gl.right_labels = False
    gl.xlabel_style = {'size': 15,}
    gl.ylabel_style = {'size': 15,}


    # plot polygons
    ab_miettemain.plot(ax=ax,color=miette_color,alpha=0.5, edgecolor='.3',label='Miette Group',zorder=50,lw=1)
    bc_miettemain.plot(ax=ax,color=miette_color,alpha=0.5, edgecolor='.3',label='Miette Group',zorder=50,lw=1)
    bc_gogmain.plot(ax=ax,color=gog_color,alpha=0.5, edgecolor='.3',label='Gog Group',zorder=50,lw=1)
    ab_gogmain.plot(ax=ax,color=gog_color,alpha=0.5, edgecolor='.3',label='Gog Group',zorder=50,lw=1)
    bc_kazamain.plot(ax=ax,color=kaza_color,alpha=0.5, edgecolor='.3',label='Kaza Group',zorder=1,lw=1)
    #bc_htmain.plot(ax=ax,color='r',alpha=0.5, edgecolor=None,label='Horsethief Creek Group',zorder=50,lw=1)
    bc_ofpmain.plot(ax=ax,color=ofp_color,alpha=1, edgecolor='.3',label='Old Fort Point Fm',zorder=50,lw=1)
    bc_byngmain.plot(ax=ax,color=byng_color,alpha=0.5, edgecolor='.3',label='Byng Formation',zorder=50,lw=1)
    bc_ismain.plot(ax=ax,color=isaac_color,alpha=0.5, edgecolor='.3',label='Isaac Fm',zorder=1,lw=1)
    ab_elsemain.plot(ax=ax,color=else_color,alpha=0.2, edgecolor='none',label='Everything Else',zorder=0)
    bc_elsemain.plot(ax=ax,color=else_color,alpha=0.2, edgecolor='none',label='Everything Else',zorder=0)

    # inset Canada
    #world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
    #canada = world[world.name=='Canada']
    axi = ax.inset_axes([0.65,0.68,0.32,0.32])
    axi.set_xticks([])
    axi.set_yticks([])
    axi.set_zorder(1000)
    cpa.plot(ax=axi,color='grey', edgecolor='.3',lw=0.25,zorder=100,alpha=0.5)
    #canada.plot(ax=axi,color='grey', edgecolor='black',lw=1,zorder=100,alpha=0.5)
    ab_geo.plot(ax=axi,color='r',alpha=1, edgecolor='r',lw=2,zorder=1000)
    bc_geo.plot(ax=axi,color='r',alpha=1, edgecolor='r',lw=2,zorder=1000)

    # plot calgary on inset for reference
    #calgary_lat = [51.045833]
    #calgary_long = [-114.0575]
    #sns.scatterplot(x=calgary_long,y=calgary_lat,marker='*',s=100,edgecolor='k',facecolor='#c8b6ff',ax=axi,zorder=1000,label='Calgary, AB')
    #axi.legend(fontsize=8,loc=3,frameon=False)

    # add field localities

    sns.scatterplot(data=data0,x='long_start', y='lat_start',marker='o',s=100,legend=False,edgecolor='k',ax=ax,zorder=1000,
                    transform=ccrs.PlateCarree(),facecolor=reds2[-4],label='Field Localities')#hue='locality',palette='Reds')
    #leg = ax.legend(bbox_to_anchor=(0.94,0.72),frameon=False,fontsize=12)
    #ax.add_artist(leg).set_zorder(1000)

    # set labels and aspect
    ax.set_xlabel('Longitude')
    ax.set_ylabel('Latitude')
    #ax.set_aspect('auto') # lets me control size

    # legend
    geos = ['Gog Gr: eC','Miette Gr: NPZ','Byng Fm: Ed','OFP Fm: Ed','Kaza Gr: NPZ','Isaac Fm: NPZ']
    geo_colors = [gog_color,miette_color,byng_color,ofp_color,kaza_color,isaac_color]
    handles = [patches.Patch(facecolor=color, label=label,alpha=0.65) for label, color in zip(geos, geo_colors)]
    #ax.legend(handles=handles,loc=3,edgecolor='k',framealpha=1,fontsize=14).set_zorder(1000)

    # North Arrow (N. is up here)
    ax.arrow(-117.35,53.49,0,0.2,width=0.125,head_length=0.2,shape='right',facecolor='grey',alpha=0.8,overhang=0.15,zorder=1000)
    ax.text(-117.567,53.3,'N',fontsize=30,family='serif',zorder=1000);

    # Label AB, and BC
    ax.text(-117,52.8,'Alberta',family='serif',transform=ccrs.PlateCarree(),zorder=1000,alpha=0.7,fontsize=18)
    ax.text(-120,52.025,'British\nColumbia',family='serif',zorder=1000,alpha=0.7,fontsize=18)

    mcb = patches.Rectangle((-120.4,53.3), .55, 0.28,edgecolor='k',facecolor='none',zorder=1000,lw=2)
    ax.add_patch(mcb)
    ax.text(-120.4,53.2,'a',family='serif',fontsize=20,zorder=1000);

    jp = patches.Rectangle((-118.4,52.7), .5, 0.25,edgecolor='k',facecolor='none',zorder=1000,lw=2)
    ax.add_patch(jp)
    ax.text(-117.82,52.875,'b',family='serif',fontsize=20,zorder=1000);

    ll = patches.Rectangle((-116.4,51.25), .5, 0.3,edgecolor='k',facecolor='none',zorder=1000,lw=2)
    ax.add_patch(ll)
    ax.text(-116.38,51.15,'c',family='serif',fontsize=20,zorder=1000);

    # add basemap
    #cx.add_basemap(ax,crs=ccrs.PlateCarree(),source=cx.providers.Stamen.Terrain,alpha=0.5,attribution=False)

    #add scalebar with function in scalebar.py
    scalebar(ax=ax,bars=1,location=(0.45, 0.075),linewidth=6,padu=1,padt=3,family='serif')
    #ax.set_aspect('auto') # lets me control size
    #ax.set_facecolor(else_color)

main_map(ax=ax,extent=main_extent,subplots=False)
```

### McKale

```python
dtm_mck = '/Users/cvanwieren/Dropbox/GitHub/OFP/OFP/Geospatial/mck_cdem_clipped.tif'
# Open the DEM with Rasterio
with rio.open(dtm_mck) as src:
    elevation_mck = src.read(1)
    elevation_mck[elevation_mck<0] = np.nan

ds_mck = gdal.Open(dtm_mck, gdal.GA_ReadOnly)
band = ds_mck.GetRasterBand(1)
L_mck = band.ReadAsArray()
[cols, rows] = L_mck.shape
ulx, xres, xskew, uly, yskew, yres = ds_mck.GetGeoTransform()
lrx = ulx + (ds_mck.RasterXSize * xres)
lry = uly + (ds_mck.RasterYSize * yres)
map_coord_extent_mck = [ulx, lrx, lry, uly]
L_mck = L_mck - 1
```
