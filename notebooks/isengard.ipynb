## From Jon for Indexing Macrostrat API

 - to see all intervals: https://macrostrat.org/api/defs/intervals?all

```python
import geopandas as gpd
import pandas as pd

#step 1: get list of unit_ids that match some search criteria, like "Neoproterozoic sediments"
units=pd.read_csv('https://macrostrat.org/api/units?interval_name=Ordovician&lith_class=sedimentary&format=csv')
#units=pd.read_csv('https://macrostrat.org/api/units?int_id=112,105&lith_class=sedimentary&format=csv')

#or "Ediacaran carbonates"
#units=pd.read_csv('https://macrostrat.org/api/units?interval_name=Ediacaran&lith_type=carbonate&format=csv')
#learn more about potential search parameters here: https://macrostrat.org/api/units

#grab the unit ids
unit_ids=[str(a) for a in units['unit_id'].tolist()]

#step 2: find map units that match these Macrostrat units
map_units=gpd.read_file('https://macrostrat.org/api/geologic_units/map?format=geojson_bare&unit_id='+','.join(unit_ids))

_=map_units.plot()
```

```python
%matplotlib inline
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
canada = world[world.name=='Canada']
import matplotlib.pyplot as plt
import cartopy.crs as ccrs

fig,ax = plt.subplots(figsize=(6,5.5))
ax = plt.axes(projection=ccrs.PlateCarree())
canada.plot(ax=ax,zorder=0)
map_units.plot(ax=ax,color='r',zorder=1000)
```
